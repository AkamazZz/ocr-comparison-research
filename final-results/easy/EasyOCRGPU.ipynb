```python
!pip install easyocr
!pip install Levenshtein
from datetime import datetime
import os
from easyocr import Reader
import Levenshtein
# Initialize the OCR reader with the desired language(s)
reader = Reader(['en'])

# Read the text from the image
photo_list = os.listdir("/content/drive/MyDrive/dataset/original/")
ground_truth = os.listdir('/content/drive/MyDrive/dataset/original groundtruth/')

photo_num = 0
accuracy_ave = 0
binarized_accuracy_ave = 0
with open('/content/drive/MyDrive/easyocr_report.txt', 'a') as f:
    f.write("=============================================================\n")
    f.write("Report on testing EasyOCR (" + str(datetime.now()) + "):\n")

for photo in photo_list:
    if photo_num == 1000:
      break
      
    text = reader.readtext('/content/drive/MyDrive/dataset/original/'+str(photo_num)+'_bn.png', detail=0)
    actual_text = ' '.join([str(elem) for elem in text])
    
    binarized_text = reader.readtext('/content/drive/MyDrive/binarized/'+str(photo_num)+'_bn.png', detail=0)
    binarized_actual_text = ' '.join([str(elem) for elem in binarized_text])

    with open('/content/drive/MyDrive/dataset/original groundtruth/'+str(photo_num)+'_bn.txt', 'r', encoding='utf-8') as f:
        lines = f.readlines()
        pre_text = str().join(lines)

    # Actual text from ground truth.
    predicted_text = pre_text
    # Calculate Levenshtein distance
    distance = Levenshtein.distance(actual_text, predicted_text)

    binarized_distance = Levenshtein.distance(binarized_actual_text, predicted_text)
    # Calculate prediction accuracy
    try:
        accuracy = 1 - (distance / len(actual_text))
        binarized_accuracy = 1 - (binarized_distance / len(binarized_actual_text))
    except Exception:
        print('Empty picture found on' + str(photo_num))
        with open('/content/drive/MyDrive/easyocr_report.txt', 'a') as f:
            f.write('Empty picture found on image num.' + str(photo_num) + ';\n')

    if photo_num % 10 == 0:
        print("Accuracy on image num. "+str(photo_num)+": " + '{"initial_image": '+str(round(accuracy * 100, 2)) + ', "binarized_image": ' + str(round(binarized_accuracy * 100, 2)) + '}' + ' percent (%);')

    accuracy_ave = accuracy_ave + accuracy
    binarized_accuracy_ave = binarized_accuracy_ave + binarized_accuracy
    with open('/content/drive/MyDrive/easyocr_report.txt', 'a') as f:
        f.write("Accuracy on image num. "+str(photo_num)+": " + '{"initial_image": '+str(round(accuracy * 100, 2)) + ', "binarized_image": ' + str(round(binarized_accuracy * 100, 2)) + '}' + ' percent (%);\n')

    photo_num = photo_num + 1


print('Average accuracy on ' + str(photo_num) + ' pictures dataset: ' +
      str(round((accuracy_ave / photo_num) * 100, 2)) + '%, on binarized image: ' +
      str(round((binarized_accuracy_ave / photo_num) * 100, 2)) + '%.')
with open('/content/drive/MyDrive/easyocr_report.txt', 'a') as f:
    f.write('Total average accuracy on ('+str(datetime.now())+') '+ str(photo_num) + ' pictures dataset: ' +
            str(round((accuracy_ave / photo_num) * 100, 2)) + '%.' + '%, on binarized image: ' +
            str(round((binarized_accuracy_ave / photo_num) * 100, 2)) + '%.\n')
```
